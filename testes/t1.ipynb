```python
import networkx as nx

# Definindo o grafo com as arestas fornecidas
G = nx.DiGraph()
edges = [
    (0, 8), (0, 6), (0, 4), (1, 6), (1, 2), (2, 9), (2, 3), (2, 8),
    (3, 5), (3, 1), (3, 4), (4, 7), (4, 5), (5, 8), (5, 7), (5, 2),
    (6, 3), (6, 0), (7, 5), (7, 2), (8, 5), (9, 3), (9, 6)
]
G.add_edges_from(edges)

# Verificando se o grafo é fortemente conectado
is_strongly_connected = nx.is_strongly_connected(G)
print(f"O grafo é fortemente conectado? {is_strongly_connected}")
```

```python
import networkx as nx
from ortools.linear_solver import pywraplp
import random 
import matplotlib.pyplot as plt 

N = random.randint(6,10)

def geraGrafo(N): 
    while True: 
        G = nx.DiGraph()
        G.add_nodes_from(range(N))

        for node in range(N):
            desc = random.randint(1,3)
            dest = random.sample([n for n in range(N) if n != node], desc)
            G.add_edges_from((node, dest) for dest in dest)

        if nx.is_strongly_connected(G):
            return G



grafo = geraGrafo(N)

for w,r in grafo.edges():
    print(f"{w} -> {r}")

nx.draw(grafo, with_labels=True, node_color='lightblue', arrows=True)
plt.show()
# output
```
